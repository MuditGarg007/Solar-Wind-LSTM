```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
import torch.nn as nn
import torch.nn.functional as F

class SolarWindDataset(Dataset):
    def __init__(self, solar_path, labels_path, seq_len=24, is_train=True):
        """
        Args:
            solar_path: Path to solar_wind.csv
            labels_path: Path to labels.csv (target dst)
            seq_len: Number of past hours to use as input (default 24 hours)
        """
        self.seq_len = seq_len
        
        # --- 1. Load Data ---
        print("Loading data...")
        # Only reading a subset of columns to save memory if needed
        features = ['bx_gse', 'by_gse', 'bz_gse', 'bt', 'density', 'speed', 'temperature']
        
        df_solar = pd.read_csv(solar_path)
        df_labels = pd.read_csv(labels_path)
        
        # Convert timedelta to timedeltas objects for easier resampling
        df_solar['timedelta'] = pd.to_timedelta(df_solar['timedelta'])
        df_labels['timedelta'] = pd.to_timedelta(df_labels['timedelta'])
        
        # --- 2. Aggregate Solar Wind Data (Minute -> Hourly) ---
        print("Aggregating solar wind data...")
        # Group by Period and Hourly Timedelta
        # We assume the timedelta is in minutes/seconds. Resample to 1H.
        # Since we have a multi-index structure (Period, Timedelta), we group by Period first.
        
        df_solar.set_index(['period', 'timedelta'], inplace=True)
        df_labels.set_index(['period', 'timedelta'], inplace=True)

        # Resample each period group to hourly means
        # Note: This step can be slow. A faster way is grouping by Period and flooring timedelta to hours.
        df_solar_hourly = df_solar.groupby([
            'period', 
            pd.Grouper(level='timedelta', freq='1h')
        ])[features].mean()
        
        # --- 3. Merge Features & Labels ---
        print("Merging and aligning...")
        # Join on index (period, timedelta)
        data = df_solar_hourly.join(df_labels, how='inner')
        
        # --- 4. Cleaning & Scaling ---
        # Interpolate missing values (common in solar wind data)
        data = data.interpolate(method='linear', limit_direction='both')
        data = data.fillna(method='ffill').fillna(method='bfill') # Fallback
        
        # Split features and target
        X = data[features].values
        y = data['dst'].values
        
        # Normalize Features
        self.scaler = StandardScaler()
        if is_train:
            X = self.scaler.fit_transform(X)
        else:
            # In a real scenario, load a saved scaler here
            X = self.scaler.fit_transform(X) 
            
        self.X = torch.FloatTensor(X)
        self.y = torch.FloatTensor(y)
        
        # Create sequences (Sliding Window)
        # We need to ensure we don't slide across different 'periods' which are non-contiguous.
        # For simplicity in this snippet, we treat the joined frame as continuous, 
        # but strictly you should group by 'period' to create sequences.
        self.sequences = self._create_sequences(data)

    def _create_sequences(self, data):
        indices = []
        # Group by period to ensure we don't create windows crossing non-continuous periods
        # 'data' still has the MultiIndex (period, timedelta)
        grouped = data.groupby(level=0) 
        
        global_idx = 0
        for name, group in grouped:
            group_len = len(group)
            if group_len > self.seq_len:
                # We can create (group_len - seq_len) sequences
                # We start at global_idx + seq_len because we need `seq_len` past steps
                start_indices = np.arange(global_idx, global_idx + group_len - self.seq_len)
                indices.extend(start_indices)
            global_idx += group_len
            
        return indices

    def __len__(self):
        return len(self.sequences)

    def __getitem__(self, idx):
        # The index 'idx' points to the START of the sequence in the flat arrays
        start = self.sequences[idx]
        end = start + self.seq_len
        
        # Input: Sequence of past 'seq_len' hours
        src = self.X[start:end]
        
        # Target: Predict t0 (current) and t+1 (next hour)
        # The 'current' time is the last step of the input sequence context
        # Kaggle goal: Predict Dst for current timestep t0 and following t+1
        
        # If our sequence is [t-23, ..., t0], then we want to predict y[t0] and y[t1]
        # However, usually "predict t0" means using data UP TO t0.
        # Let's assume we use data [t-24 ... t-1] to predict t0 and t+1.
        
        target_t0 = self.y[end]       # Current hour
        # Check boundary for t+1
        if end + 1 < len(self.y):
            target_t1 = self.y[end + 1]
        else:
            target_t1 = target_t0 # Fallback at very end
            
        target = torch.tensor([target_t0, target_t1])
        
        return src, target
```

```python
class SolarAttentionLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim=2, dropout=0.2):
        super(SolarAttentionLSTM, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        
        # LSTM Layer
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout
        )
        
        # Attention Layer
        self.attention_fc = nn.Linear(hidden_dim, 1)
        
        # Output Layer
        self.fc = nn.Linear(hidden_dim * 2, output_dim)
        
    def forward(self, x, return_attention=False):
        # 1. Run LSTM
        lstm_out, (hidden, cell) = self.lstm(x)
        
        # 2. Calculate Attention
        scores = self.attention_fc(lstm_out)
        attention_weights = F.softmax(scores, dim=1)
        
        # 3. Context Vector
        context_vector = torch.sum(lstm_out * attention_weights, dim=1)
        
        # 4. Prediction
        last_hidden = lstm_out[:, -1, :]
        combined = torch.cat((context_vector, last_hidden), dim=1)
        out = self.fc(combined)
        
        if return_attention:
            return out, attention_weights
        else:
            return out
```

```python
# --- Hyperparameters ---
SEQ_LEN = 24       # Use past 24 hours to predict
BATCH_SIZE = 64
HIDDEN_DIM = 128
NUM_LAYERS = 2
LEARNING_RATE = 0.001
EPOCHS = 100

# --- Setup Data ---
# Replace paths with your actual Kaggle paths
# e.g., '/kaggle/input/solar-wind/solar_wind.csv'
SOLAR_PATH = 'archive/solar_wind.csv'
LABELS_PATH = 'archive/labels.csv'

# Note: Loading dataset might take time due to aggregation
full_dataset = SolarWindDataset(SOLAR_PATH, LABELS_PATH, seq_len=SEQ_LEN)

# Simple Train/Val split (80/20)
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
torch.manual_seed(42)
train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Setup Model ---
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
input_dim = 7  # Number of features selected in Dataset class
model = SolarAttentionLSTM(input_dim=7, hidden_dim=128, num_layers=2).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

# --- Training Loop ---
print("Starting training...")
for epoch in range(EPOCHS):
    model.train()
    train_loss = 0
    
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
        
    # Validation
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for X_val, y_val in val_loader:
            X_val, y_val = X_val.to(device), y_val.to(device)
            preds = model(X_val)
            v_loss = criterion(preds, y_val)
            val_loss += v_loss.item()
            
    avg_train_loss = train_loss / len(train_loader)
    avg_val_loss = val_loss / len(val_loader)
    
    print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

# --- Inference Example ---
# To predict on new data, ensure it is preprocessed identically (scaled/aggregated)
# sample_input = torch.randn(1, 24, 7).to(device) # (Batch, Seq_Len, Features)
# prediction = model(sample_input)
# print("Predicted Dst (t0, t+1):", prediction)
```

```python
import matplotlib.pyplot as plt

def plot_predictions(model, val_loader, device):
    model.eval()
    actuals = []
    predictions = []
    
    with torch.no_grad():
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(device)
            # Forward pass
            out = model(X_batch)
            
            # Append predictions and actuals
            # We focus on t0 (the first output column) for the plot
            predictions.extend(out[:, 0].cpu().numpy())
            actuals.extend(y_batch[:, 0].cpu().numpy())

    # Plot specific range (e.g., first 200 hours) to see details
    plt.figure(figsize=(12, 6))
    plt.plot(actuals[:200], label='Actual Dst', color='black', alpha=0.7)
    plt.plot(predictions[:200], label='Predicted Dst', color='red', linestyle='--')
    plt.title('Solar Wind Prediction: Actual vs Predicted (First 200 hrs of Val Set)')
    plt.xlabel('Time (Hours)')
    plt.ylabel('Dst Index (nT)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the visualization
plot_predictions(model, val_loader, device)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import torch

def plot_attention_map(model, loader, device):
    model.eval()
    
    # Get one batch of data from validation set
    X_batch, y_batch = next(iter(loader))
    X_batch = X_batch.to(device)
    
    # Forward pass with the new flag
    with torch.no_grad():
        # returns (predictions, weights) because flag is True
        preds, attention_weights = model(X_batch, return_attention=True)
        
        # attention_weights shape: (Batch_Size, Seq_Len, 1)
        # We squeeze it to (Batch_Size, Seq_Len) for plotting
        weights = attention_weights.squeeze(-1).cpu().numpy()
        
    # Pick the first 15 samples in the batch to display
    num_samples_to_plot = 15
    weights_subset = weights[:num_samples_to_plot, :]

    # Plotting
    plt.figure(figsize=(12, 8))
    sns.heatmap(weights_subset, cmap='inferno', xticklabels=True, yticklabels=False)
    
    plt.title("Attention Heatmap: Which Past Hour Matters Most?", fontsize=16)
    plt.xlabel("Hours in Past Window (0 = Oldest, 23 = Recent)", fontsize=12)
    plt.ylabel(f"First {num_samples_to_plot} Test Samples", fontsize=12)
    plt.show()

# Run the visualization
plot_attention_map(model, val_loader, device)
```
